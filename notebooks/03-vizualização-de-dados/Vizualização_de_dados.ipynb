##Visualização #01: Gráfico de Barras da Arrecadação de Filmes

```python
#Importando as bibliotecas:

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Carregando os dados:

url = 'https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv'
df = pd.read_csv(url)
df.head()
```

```python
#Criando o gráfico de barras

df['arrecadacao_milhao'] = df['arrecadacao'] / 1000000
df = df.sort_values(by='arrecadacao', ascending=False)
sns.set_style("whitegrid")

plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='center', color='white')

plt.xlabel('Arrecadação (Milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')
ax.set(frame_on=False)
ax.axvline(0, color="k", clip_on=False)
plt.show()
```

##Visualização #02: Histograma da Massa dos Pinguins

```python
#Importando as bibliotecas e carregando os dados:

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = sns.load_dataset("penguins")
df.head()
```

```python
#Criando o histograma

sns.set_style("whitegrid")
ax = sns.histplot(data=df, x='body_mass_g')
plt.xlabel('Massa (g)')
plt.ylabel('Quantidade')
plt.title('Distribuição da massa dos pinguins')
ax.set(frame_on=False)
ax.axhline(0, color="k", clip_on=False)
plt.show()
```

##Visualização #03: Boxplot da Quantidade Total da Conta em Relação ao Dia da Semana

```python
#Importando as bibliotecas e carregando os dados:

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = sns.load_dataset("tips")
df.head()
```

```python
#Criando o boxplot

sns.set_style("whitegrid")
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x="day", y="total_bill", data=df, color='cornflowerblue')
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
